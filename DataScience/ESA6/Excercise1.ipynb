RAW dataset:

```python
import pandas as pd
import numpy as np
import requests

url = "https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv"
raw_df = pd.read_csv(url)
raw_df
```

Remove empty rows:

```python
raw_df.dropna(how="all", inplace=True)
raw_df
```

Search for duplicates in name:

```python
raw_df.duplicated("full_name")
```

Search for duplicates in email:

```python
raw_df.duplicated("email")
```

Remove duplicates (Eden Wace)

```python
raw_df.drop_duplicates(subset="email", keep="first", inplace=True)
raw_df
```

Remove rows with missing email and gender informationa because its not possible to replace with valid information

```python
raw_df.dropna(subset=["email", "gender"], inplace=True)
raw_df
```

Clean age entries (remove all non-numeric entries, negative values are interpreted as typos and get absolute values)

```python
raw_df['age'] = pd.to_numeric(raw_df['age'], errors='coerce')
raw_df.dropna(subset=['age'], inplace=True)
raw_df['age'] = raw_df['age'].astype('int')
raw_df['age'] = raw_df['age'].abs()
raw_df
```

Clean ids (cast id to int, find largest entry, NaN ids get the next possible id)

```python
raw_df['id'] = pd.to_numeric(raw_df['id'], errors='coerce')
new_max_id = raw_df['id'].max() + 1
raw_df.id = raw_df.id.fillna(new_max_id + (raw_df.id.isnull().cumsum()-1))
raw_df['id'] = raw_df['id'].astype('int')
raw_df
```
